```python
from datetime import datetime, timedelta
import pandas as pd

# Function to calculate maintenance due date for each row
def calculate_maintenance_due_date(row):
    installation_date = pd.to_datetime(row['Installation_Date'])
    equipment_type = row['Equipment_Type']
    acquisition_type = row['Acquisition_Type']
    last_maintenance_date = pd.to_datetime(row['Last_Date_Maintained'])
    maintenance_interval = row['Maintenance_Interval']

    # Check if maintenance interval is NaN
    if pd.isna(maintenance_interval):
        return pd.NaT, pd.NaT

    # Extract the numeric part from maintenance interval
    maintenance_interval = int(''.join(filter(str.isdigit, str(maintenance_interval))))

    # Calculate initial due date
    initial_due_date = installation_date + timedelta(days=maintenance_interval * 30)

    # If last maintenance date is available and valid, use it instead
    if pd.notnull(last_maintenance_date):
        last_due_date = last_maintenance_date + timedelta(days=maintenance_interval * 30)
        if last_due_date > initial_due_date:
            initial_due_date = last_due_date

    # Calculate maintenance due date for 2024
    maintenance_due_date_2024 = initial_due_date.replace(year=2024)

    return initial_due_date, maintenance_due_date_2024

# Read the CSV file
df = pd.read_csv('Installation_Note_April.csv')

# Apply the function to calculate maintenance due dates for 2024
df[['Initial_Due_Date', 'Maintenance_Due_Date_2024']] = df.apply(calculate_maintenance_due_date, axis=1, result_type='expand')

# Print the dataframe with selected columns
# Select columns for the DataFrame
selected_columns = ['Installation_Date', 'Customer', 'Item_Code', 'Equipment_Type', 
                    'Acquisition_Type', 'Last_Date_Maintained', 'Initial_Due_Date', 
                    'Maintenance_Due_Date_2024']

# Create a new DataFrame with selected columns
selected_df = df[selected_columns]

# Print the DataFrame
print(selected_df)


# Save the dataframe to a CSV file
df.to_csv('maintenance_due_dates.csv', index=False)

print("\nDataFrame saved to maintenance_due_dates.csv")
```

```python
# Save the DataFrame to a CSV file
results_df.to_csv('maintenance_schedule.csv', index=False)

print("\nDataFrame saved to maintenance_schedule.csv")
```

```python
from datetime import datetime, timedelta
import pandas as pd

# Read the CSV file
df = pd.read_csv('Contractor_Installation_Note.csv')

# Define function to calculate maintenance due date based on equipment type
def calculate_maintenance_due_date(installation_date, equipment_type):
    if equipment_type == 'Analyzer':
        # Maintenance due in 6 months for analyzers
        return installation_date + timedelta(days=6 * 30)
    else:
        # Maintenance due in 12 months for non-analyzers
        return installation_date + timedelta(days=12 * 30)

# Initialize an empty list to store results
results = []

# Define sites and corresponding columns
sites = ['KIBI', 'TQMH', 'ABURI']
site_columns = {
    'KIBI': ['ITEM DESCRIPTION', 'Equipment_Type', 'KIBI', 'Installation_Date'],
    'TQMH': ['ITEM DESCRIPTION', 'Equipment_Type', 'TQMH', 'Installation_Date'],
    'ABURI': ['ITEM DESCRIPTION', 'Equipment_Type', 'ABURI', 'Installation_Date']
}

# Iterate over each site
for site in sites:
    # Extract columns related to the current site
    site_df = df[site_columns[site]].dropna(subset=[site])
    
    # Iterate over each row of the site DataFrame
    for index, row in site_df.iterrows():
        # Extract equipment details
        equipment_description = row['ITEM DESCRIPTION']
        equipment_type = row['Equipment_Type']
        installation_date = pd.to_datetime(row['Installation_Date'])
        
        # Calculate maintenance due date
        maintenance_due_date = calculate_maintenance_due_date(installation_date, equipment_type)
        
        # Append the results to the list
        results.append({
            'Site': site,
            'Equipment_Description': equipment_description,
            'Equipment_Type': equipment_type,
            'Installation_Date': installation_date,
            'Maintenance_Due_Date': maintenance_due_date
        })

# Concatenate the results list into a DataFrame
results_df = pd.DataFrame(results)

# Print the results DataFrame
print(results_df)

# Save the DataFrame to a CSV file
results_df.to_csv('maintenance_schedule.csv', index=False)

print("\nDataFrame saved to maintenance_schedule.csv")

```
